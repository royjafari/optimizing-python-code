### What is a tree?

```python
class TreeNode(object):
    def __init__(self,val=None):
        self.val = val
        self.children = []
```

```python
n1 = TreeNode(1)
n2 = TreeNode(2)
n3 = TreeNode(3)
n4 = TreeNode(4)
n5 = TreeNode(5)
n6 = TreeNode(6)
n7 = TreeNode(7)

n1.children.extend([n2,n3])
n2.children.extend([n4,n5,n6])
n3.children.append(n7)
```

```python
def print_tree_from_root(root):
    
    print_order = []
    def get_print_order(root,l=0):
        if len(root.children) ==0:
            return (root.val,l)
        print_order.append((root.val,l))
        l +=1
        for child in root.children:
            print_order.append(get_print_order(child,l))
            
    get_print_order(root)
    
    
    for val,l in [tup for tup in print_order if tup]:
        print('___'*l + str(val))
```

```python
print_tree_from_root(n1)
```

### What is a Graph?
#### List of Edges

```python
import pandas as pd
class GraphLE(object):
    def __init__(self,n_nodes=None):
        self.edge_df = pd.DataFrame(
            columns = ['n1','n2','w']
        )
        self.n_nodes = n_nodes
    def add_edge(self,n1,n2,w=1):
        add_df = pd.DataFrame(
            {'n1': n1, 'n2': n2, 'w': w},
            index=[len(self.edge_df)]
        )
        self.edge_df = pd.concat([self.edge_df,add_df])
    def print_all_adj_list(self):
        print(self.edge_df)
    def get_edge_from(self,node):
        BM = self.edge_df.n1 == node
        return [(r.n1,r.n2,r.w) for i,r in 
               self.edge_df[BM].iterrows()]
```

```python
le_graph = GraphLE(7)
edge_list = [(3,0,5),(3,1,2),(1,0,7),(0,2,1),(2,5,9),(6,2,3),(5,6,3),(5,1,2),(5,4,9),(4,1,3),(4,5,8)]
for n1,n2,w in edge_list:
    le_graph.add_edge(n1,n2,w)
```

```python
le_graph.print_all_adj_list()
```

```python
print(le_graph.get_edge_from(5))
```

#### Adjecancy Matrix

```python
import numpy as np
import pandas as pd
class GraphAM(object):
    def __init__(self,n_nodes):
        self.n_nodes = n_nodes
        self.adj_matrix = (
            np.zeros([n_nodes,n_nodes])
            .astype(int)
        )
    def add_edge(self,n1,n2,w=1):
        self.adj_matrix[n1,n2] = w
    def print_adj_matrix(self):
        print(self.adj_matrix)
    def get_edge_from(self,n1):
        return [(n1,n2,w) for n2,w in
                enumerate(self.adj_matrix[n1,:]) if w>0]
```

```python
am_graph = GraphAM(7)
edge_list = [(3,0,5),(3,1,2),(1,0,7),(0,2,1),(2,5,9),(6,2,3),(5,6,3),(5,1,2),(5,4,9),(4,1,3),(4,5,8)]
for n1,n2,w in edge_list:
    am_graph.add_edge(n1,n2,w)
```

```python
am_graph.print_adj_matrix()
```

```python
print(am_graph.get_edge_from(5))
```

```python
class TreeNode(object):
    def __init__(self,name=None,age=None):
        self.left = None
        self.right = None
        self.name = name
        self.age = age
```

```python
def insert(root,new_name,new_age):
    if root is None:
        root=TreeNode(new_name,new_age)
        return root
 
    if new_name<root.name:
        root.left=insert(root.left,new_name,new_age)
    else:
        root.right=insert(root.right,new_name,new_age)
    return root
```

```python
import pandas as pd
import numpy as np

def dijkstra(graph,source):
    n_nodes = graph.n_nodes
    
    min_dist = pd.Series(float('inf'),index=range(n_nodes))
    min_dist[source] = 0
    
    finalized = pd.Series(False,index=range(n_nodes))
    final_path = {node:None for node in range(n_nodes)}
    
    while sum(finalized) <n_nodes:
        node = min_dist[finalized==False].idxmin()
        for n1,n2,w in graph.get_edge_from(node):
            
            new_w = min_dist.iloc[n1] + w 
            
            if(new_w< min_dist.iloc[n2]):
                min_dist.iloc[n2] = new_w
                final_path[n2] = n1
        
        finalized[node] = True
        
    shortest_paths = {}
    
    for i in range(n_nodes):
        path = []
        dest = i

        path.append(dest)

        while final_path[dest] != None:
            path.append(final_path[dest])
            dest = final_path[dest]
        
        shortest_paths[i] = path[::-1]

    return min_dist, shortest_paths
```
